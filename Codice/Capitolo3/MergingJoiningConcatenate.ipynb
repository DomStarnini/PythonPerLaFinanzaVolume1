```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3,4],
                 'B':[5,6,7,8],
                 'C':[9,10,11,12],
                 'D':[13,14,15,16]})
```

```python
df2=pd.DataFrame({'A':[17,18,19,20],
                 'B':[21,22,23,24],
                 'C':[25,26,27,28],
                 'D':[29,30,31,32]})
```

```python
df3=pd.DataFrame({'A':[33,34,35,36],
                 'B':[37,38,39,40],
                 'C':[41,42,43,44],
                 'D':[45,46,47,48]})
```

```python
df4=pd.DataFrame({'A':[1,2,3,4],
                 'B':[5,6,7,8],
                 'C':[9,10,11,12],
                 'D':[13,14,15,16],
                 'K':['A','B','C','D']})
```

```python
df5=pd.DataFrame({'A':[17,18,19,20],
                 'B':[21,22,23,24],
                 'C':[25,26,27,28],
                 'D':[29,30,31,32],
                 'K':['A','B','C','D']})
```

```python
pd.merge(df4,df5,how='inner',on='K')
```

```python
df6=pd.DataFrame({'A':[1,2,3,4],
                 'B':[5,6,7,8],
                 'C':[9,10,11,12],
                 'D':[13,14,15,16]},
                 index=['A','B','C','D'])
```

```python
df7=pd.DataFrame({'E':[17,18,19,20],
                 'F':[21,22,23,24],
                 'G':[25,26,27,28],
                 'H':[29,30,31,32]},
                 index=['A','F','E','A'])
```

```python
df6.join(df7)
```
