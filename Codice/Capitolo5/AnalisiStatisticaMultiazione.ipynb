```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import date
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates

pd.plotting.register_matplotlib_converters()
```

```python
titoli = ["FB", "AMZN", "AAPL", "NFLX", "GOOG","MSFT"]
inizio=date(2014,1,1)
fine=date.today()
data=yf.download(titoli,inizio,fine)
df=pd.DataFrame(data)
df.head()
```

```python
fig = plt.figure(figsize=(18, 12))
fig.set_facecolor('white')
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)
ax1.grid(color='black')
ax1.plot(df['Close']['AMZN'])
ax1.set_title("Amazon")
ax2.grid(color='black')
ax2.plot(df['Close']['AAPL'])
ax2.set_title("Apple")
ax3.grid(color='black')
ax3.plot(df['Close']['FB'])
ax3.set_title("Facebook")
ax4.grid(color='black')
ax4.plot(df['Close']['NFLX'])
ax4.set_title("Netflix")
ax5.grid(color='black')
ax5.plot(df['Close']['GOOG'])
ax5.set_title("Google")
ax6.grid(color='black')
ax6.plot(df['Close']['MSFT'])
ax6.set_title("Microsoft")
plt.show()
```

```python
multi_titoli_ritorno_giornaliero = df['Close'].pct_change()
multi_titoli_ritorno_giornaliero.head()
```

```python
multi_titoli_ritorno_giornaliero.describe()
```

```python
fig = plt.figure(figsize=(18, 12))
ax1 = fig.add_subplot(111)
ax1.grid(color='black')
ax1.plot(multi_titoli_ritorno_giornaliero)
ax1.set_xlabel("Data")
ax1.set_ylabel("Percentuale")
ax1.set_title("Ritorno Giornaliero")
ax1.legend(titoli)
plt.show()
```

```python
fig = plt.figure(figsize=(18, 12))
ax1 = fig.add_subplot(111)
ax1.grid(color='black')
ax1.plot((multi_titoli_ritorno_giornaliero + 1).cumprod())
ax1.legend(titoli)
plt.show()
```
