```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_da_importare='FTSE MIB Dati Storici.csv'
def importaCSV(fileCSV):
  data=pd.read_csv(fileCSV,usecols=['Data','Ultimo','Apertura','Massimo','Minimo'])
  data_finale=data['Data'].astype(str).str.replace('.','-')
  close=data['Ultimo'].astype(str).str.replace('.','').str.replace(',','.').astype(np.float)
  open=data['Apertura'].astype(str).str.replace('.','').str.replace(',','.').astype(np.float)
  high=data['Massimo'].astype(str).str.replace('.','').str.replace(',','.').astype(np.float)
  low=data['Minimo'].astype(str).str.replace('.','').str.replace(',','.').astype(np.float) 
  data_dict={'data':data_finale,'close':close,'open':open,'high':high,'low':low}
  df=pd.DataFrame.from_dict(data_dict)
  df.set_index('data',inplace=True) 
  return df

df=importaCSV(file_da_importare)
print(df)
  
```

```python
plt.figure(num=None, figsize=(18, 12))
plt.xticks(rotation='vertical')
plt.yticks(rotation='horizontal')
plt.xlabel('Data',size=35,fontweight='bold')
plt.ylabel('Prezzo',size=35,fontweight='bold')
plt.plot(df['close'],label='FTSE MIB')
plt.grid(True,linewidth=0.3,color='#000000',linestyle='-')
plt.legend( ['FTSE MIB'])
plt.title('INVESTING CSV DATA FTSE MIB',size=35,fontweight='bold')
plt.show()
```
