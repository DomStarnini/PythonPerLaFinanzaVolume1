```python
plt.xticks(rotation='vertical')
plt.plot(df['Date'],df['Close'])
plt.show()
```

```python
plt.xlabel('Data',size=15,fontweight='bold')
plt.ylabel('Prezzo',size=15,fontweight='bold')
plt.xticks(rotation='vertical')
plt.plot(df['Date'],df['Close'])
plt.show()
```

```python
plt.grid(True,linewidth=0.5,color='#000000',linestyle='-')
plt.xlabel('Data',size=15,fontweight='bold')
plt.ylabel('Prezzo',size=15,fontweight='bold')
plt.xticks(rotation='vertical')
plt.plot(df['Date'],df['Close'])
plt.show()
```

```python
plt.title("BTC_USD",size=20,fontweight='bold')
plt.grid(True,linewidth=0.5,color='#000000',linestyle='-')
plt.xlabel('Data',size=15,fontweight='bold')
plt.ylabel('Prezzo',size=15,fontweight='bold')
plt.xticks(rotation='vertical')
plt.plot(df['Date'],df['Close'])
plt.show()
```

```python
plt.title("BTC_USD",size=20,fontweight='bold')
plt.grid(True,linewidth=0.5,color='#ff0000',linestyle='-')
plt.xlabel('Data',size=15,fontweight='bold')
plt.ylabel('Prezzo',size=15,fontweight='bold')
plt.xticks(rotation='vertical')
plt.plot(df['Date'],df['Close'],label='PREZZO BTC-USD')
plt.legend(loc=3)
plt.show()
```
