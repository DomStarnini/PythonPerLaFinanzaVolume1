```python
for x in range(50):
  y=np.random.randint(50, size=1) 
  z=np.random.randint(50, size=1)
  k=np.random.randint(50, size=1)  
  plt.plot(y, x, marker='o', markersize=9, color="red")
  plt.plot(z,x, marker='*', markersize=9, color="green")
  plt.plot(k,x, marker='+', markersize=9, color="blue")
plt.show()
```

```python
nomi = 'AAPL', 'MSFT', 'BTC', 'ETH'
sizes = [15, 30, 45, 10]
fig1,ax1 = plt.subplots()
ax1.pie(sizes, labels=nomi)
plt.show()
```

```python
nomi = 'AAPL', 'MSFT', 'BTC', 'ETH'
sizes = [15, 30, 45, 10]
fig1,ax1 = plt.subplots()
ax1.barh(nomi, sizes, align='center',color='green', ecolor='black')
plt.show()
```

```python
sizes1 = np.random.randint(100, size=500)
sizes = np.random.randint(50, size=500)
fig1,ax1 = plt.subplots()
ax1.scatter(sizes1, sizes)
plt.show()
```

```python
sizes = np.random.randint(50, size=500)
fig1,ax1 = plt.subplots()
ax1.hist(sizes, bins=15, density=True)
plt.show()
```
